```python
import numpy as np
X=np.load('/content/mouse_movement_features.npy')
Y=np.load('/content/mouse_movement_labels.npy')
```

```python
from tensorflow.keras.layers import Input,Dense,Bidirectional,LSTM
from tensorflow.keras.models import Model,load_model
from tensorflow.keras.optimizers import Adam
```

```python

inputs=Input((100,9))
step=Bidirectional(LSTM(6))(inputs)
pred=Dense(2)(step)
model=Model(inputs=inputs,outputs=pred)
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])
model.summary()
```

```python

#X=np.array(user)[:,:,:-1]
#Y=np.zeros([X.shape[0],1])
model.fit(x=X,y=Y,batch_size=16,epochs=1,validation_split=0.1)
```

```python
#model=load_model('/content/mouse_model.h5')
pred=model.predict(X[593][np.newaxis,:,:])
pred[0][0]
```
